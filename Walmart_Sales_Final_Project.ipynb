# Which store has maximum sales?

### use groupby() ,agg() funcation and choose sum operate to do in Weekly_Sales column.

```python
total_Weekly_Sales_for_store = WalmartData.groupby(['Store'], as_index=False).agg(Weekly_Sales=('Weekly_Sales','sum')).round()
print(total_Weekly_Sales_for_store.head())
```

### return Maximum weekly_Sales and Store Number

```python
total_Weekly_Sales_for_store[(total_Weekly_Sales_for_store['Weekly_Sales'] == max(total_Weekly_Sales_for_store["Weekly_Sales"]))]
```

## visulaise our output to get a good look 

```python
# Scatter Plot 
plt.figure(figsize = (20, 7))
plt.style.use('seaborn')
X = total_Weekly_Sales_for_store["Store"]
Y = total_Weekly_Sales_for_store["Weekly_Sales"]
plt.xticks(np.arange(46))
plt.scatter(X, Y, s=200, c='g')
plt.title("Sales distribution according to stores")
plt.xlabel("Store Number")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(20,10))
MaxSales=WalmartData.groupby(['Store'])
MaxSales.Weekly_Sales.sum().sort_values(ascending=False).plot.bar(color='Red')
plt.xlabel('Store number')
plt.ylabel('Sales')
plt.title('Sales distribution according to stores' , color='b')
plt.show()
```

# Some holidays have a negative impact on sales. Find out holidays 
# that have higher sales than the mean sales in the non-holiday season 
# for all stores together.

```python
WalmartData["Date"] = pd.to_datetime(WalmartData.Date, format="%d-%m-%Y")
WalmartData['Month'] = pd.to_datetime(WalmartData['Date'], format="%d-%m-%Y").dt.month


conditions=[
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==2),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==9),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==11),
    (WalmartData['Holiday_Flag']==1) & (WalmartData['Month']==12),
    (WalmartData['Holiday_Flag']==0)
]
choices=['Super Bowl','Labour Day','Thanksgiving','Christmas','Non Holidays']
holiday_analysis_df = WalmartData.drop(['Store', 'Month'], axis='columns')
holiday_analysis_df['Holiday']=np.select(conditions,choices)
mean_df = holiday_analysis_df.groupby('Holiday').agg(Mean_Weekly_Sales=('Weekly_Sales','mean')).round()
mean_df.plot(kind='bar', y='Mean_Weekly_Sales', figsize=(10,10))
mean_df.sort_values(by='Mean_Weekly_Sales',ascending = False)
```

# Provide a monthly and semester view of sales in units and give insights.


```python
#Monthly sales 
monthly = WalmartData.groupby(pd.Grouper(key='Date', freq='1M')).sum()# groupby each 1 month
monthly=monthly.reset_index()
fig, ax = plt.subplots(figsize=(10,8))
X = monthly['Date']
Y = monthly['Weekly_Sales']
plt.plot(X,Y)
plt.title('Month Wise Sales')
plt.xlabel('Monthly')
plt.ylabel('Weekly_Sales')
plt.show()
```

# Plot the relations between weekly sales vs. other numeric features and give insights.

### relation between weekly sales and Stores number.

```python
plt.figure(figsize=(20,10))
MaxSales=WalmartData.groupby(['Store'])
MaxSales.Weekly_Sales.sum().sort_values(ascending=False).plot.bar(color='Red')
plt.xlabel('Store number')
plt.ylabel('Sales')
plt.title('Sales distribution according to stores' , color='b')
plt.show()
```

## - we find that store 20 has maximum sales and store 33 has minimum sales .

### relation between weekly sales per year.

```python
WalmartData['Year'] = pd.to_datetime(WalmartData['Date'], format="%d-%m-%Y").dt.year
annually_df = WalmartData.groupby(['Year'], as_index=False).agg(Weekly_Sales=('Weekly_Sales','sum'))  
sns.barplot(x = 'Year',y = 'Weekly_Sales',data = annually_df)
plt.title('Weekly Sales per Year')
plt.xlabel('Year')
plt.ylabel('Weekly Sales')
plt.show()
```

## - we find that 2011 year has maximum weekly sales .

### relation between weekly sales per Monthes.

```python
WalmartData['month_year'] = pd.to_datetime(WalmartData['Date']).dt.to_period('M')

monthly_df = WalmartData.groupby(['month_year'], as_index=False).agg(Weekly_Sales=('Weekly_Sales','sum'))
monthly_df.head()
fig, ax = plt.subplots(figsize=(30,5))
sns.barplot(x = 'month_year',y = 'Weekly_Sales',data = monthly_df,ax=ax)
plt.title('Weekly Sales each month of the year')
plt.xlabel('Months')
plt.ylabel('Weekly Sales')
plt.show()
```

## - We find (2010-12) and (2011-12) have Maximum Sales .

### relation between weekly sales per Semester.

```python
#Semester Sales 
Semester = WalmartData.groupby(pd.Grouper(key='Date', freq='6M')).sum()
Semester = Semester.reset_index()
fig, ax = plt.subplots(figsize=(10,8))
X = Semester['Date']
Y = Semester['Weekly_Sales']
plt.plot(X,Y)
plt.title('Semester Wise Sales')
plt.xlabel('Semester')
plt.ylabel('Weekly_Sales')
plt.show()
```
